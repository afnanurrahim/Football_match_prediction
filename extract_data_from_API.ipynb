## Getting match statistics

```python
url = "https://api-football-beta.p.rapidapi.com/fixtures/statistics"
querystring = {"fixture": str(867946)}
response_stats = requests.get(url, headers=headers, params=querystring)
```

```python
response_stats.json()
```

```python
home_cols = ['fixture id', 'Home team id', 'Home team name', 'Shots on Goal',
       'Shots off Goal', 'Total Shots', 'Blocked Shots', 'Shots insidebox',
       'Shots outsidebox', 'Fouls', 'Corner Kicks', 'Offsides',
       'Ball Possession', 'Yellow Cards', 'Red Cards', 'Goalkeeper Saves',
       'Total passes', 'Passes accurate', 'Passes %']

away_cols= ['fixture id','away team id', 'away team name', 'Shots on Goal', 'Shots off Goal',
       'Total Shots', 'Blocked Shots', 'Shots insidebox', 'Shots outsidebox',
       'Fouls', 'Corner Kicks', 'Offsides', 'Ball Possession', 'Yellow Cards',
       'Red Cards', 'Goalkeeper Saves', 'Total passes', 'Passes accurate',
       'Passes %', 'away team id', 'away team name', 'expected_goals']
```

```python
home_df = pd.DataFrame(columns=cols)
away_df= pd.DataFrame(columns=cols)
```

```python
url = "https://api-football-beta.p.rapidapi.com/fixtures/statistics"

for fixture_id in df['fixture.id'][291:]:
    querystring = {"fixture": str(fixture_id)}
    response_stats = requests.get(url, headers=headers, params=querystring)
    
    match_info = response_stats.json().get('response')

    home={}
    home['fixture id'] = fixture_id
    home['Home team id'] = match_info[0]['team']['id']
    home['Home team name'] = match_info[0]['team']['name']

    for r in match_info[0]['statistics']:
        home[r['type']] = r['value']

    home_df = pd.concat([home_df,pd.DataFrame([home])])

    away={}
    away['fixture id'] = fixture_id
    away['away team id'] = match_info[1]['team']['id']
    away['away team name'] = match_info[1]['team']['name']

    for r in match_info[1]['statistics']:
        away[r['type']] = r['value']

    away_df = pd.concat([away_df,pd.DataFrame([away])])
```

```python
response_stats.json()
```

```python
home_df=home_df.set_index('fixture id')
away_df=away_df.set_index('fixture id')
```

```python
away_df.to_csv("2023_away_teams_stats.csv")
```
